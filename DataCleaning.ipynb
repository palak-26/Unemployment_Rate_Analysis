```python
# Removing rows with missing values
df_cleaned = df.dropna()

# Displaying the DataFrame after removing missing values
print("\nDataFrame after removing rows with missing values:")
print(df_cleaned)
```

```python
sns.boxplot(df_cleaned[' Estimated Unemployment Rate (%)'])
```

```python
def removal_box_plot(df, column, threshold):
    sns.boxplot(df[column])
    plt.title('Original Box Plot ')
    plt.show()

    removed_outliers = df[df[column] <= threshold]

    sns.boxplot(removed_outliers[column])
    plt.title(f'Box Plot without Outliers')
    plt.show()
    return removed_outliers


threshold_value = 1

no_outliers = removal_box_plot(df_cleaned, ' Estimated Unemployment Rate (%)', threshold_value)
```
